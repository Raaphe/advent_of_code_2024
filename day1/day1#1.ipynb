### Load File Input
I will load the input file.

```python

file = None;
with open("./input.txt", "r") as file:
    file = [line.strip() for line in file];
```

Removed duplicates and put them in seperate sorted lists.

```python
fileList = [line.split("   ") for line in file];

set1 = sorted([line[0] for line in fileList]);
set2 = sorted([line[1] for line in fileList]);   

set1
```

```python
res = 0;

for i in range(len(set1)):
    res += abs(int(set1[i]) - int(set2[i]));

res
```
